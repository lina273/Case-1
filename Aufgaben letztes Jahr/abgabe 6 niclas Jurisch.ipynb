# Servomotor-Antriebe

Schreiben Sie eine Klasse für Sensoren mit den folgenden Attributen und implementieren sie auch die Methoden. Instanziieren Sie zwei Servomotor-Antriebe Ihrer Wahl und fügen Sie diesem einem `RobotController` hinzu und testen Sie die Methoden.

- Verfassen Sie eine Klasse für die Servomotor-Antriebe `ServoMotor` eines Roboter-Arms mit den Attributen Namen, Soll-Winkel, Ist-Winkel, und Leistungs-Aufnahme (welche zunächst auf 0 gesetzt wird). Jeder Servomotor hat eine Methode `show_current_angle()` und `show_target_angle()`, um die Winkel mittels `print()`-Befehl auszugeben. 
- Zudem wird eine Setter-Methode `set_target_angle()` implementiert, um das Attribut für den Soll-Winkel zu setzen. Immer wenn der Soll-Winkel angepasst wird, so wird auch die  Methode `show_target_angle()` aufgerufen, um die Nutzer:in von der Änderung zu informieren.
- Zudem gibt es eine Klasse `RobotController`. Der `RobotController` hat ebenfalls einen Namen und kann in einem weiteren Attribut `motors` eine Liste von Objekten der Klasse `ServoMotor` enthalten. Ebenso enthält er eine Liste `sensors` von Objekten der `Sensor`-Klasse aus der letzten Hausübung.
- Zudem hat der `RobotController` eine Methode `activate_resting_position()` mit der er alle in ihm verbauten Motoren auf einen Winkel von `0` Grad setzt.

Verfassen Sie auch DocStrings und TypeHints.

```python
# Import der Sensor Klasse
class Sensor():
    def __init__(self, id, sensorname, measurement):
        """
        Initialisierung Sensordaten

        :param id: 
        :param sensorname: 
        :param measurements: 
        """
        self.id: int = id
        self.sensorname: str = sensorname
        self.measurement: list[float] = measurement
        self.last_measurement: float = 0.0

    def set_last_measurement(self, measurement):
        """
        Setzen des letzten Wertes
        :param messwert: 
        :return: 
        """
        self.last_measurement = measurement
        
        if len(self.measurement) < 10:
            self.measurement.append(self.last_measurement)
        else:
            del self.measurement[0]
            self.measurement.append(measurement)

    def calculate_mean(self):
        """
        Berechnung
        :return: 
        """
        self.mean = sum(self.measurement) / len(self.measurement)       
        
    def print_mean(self):
        """
        Ausgabe Berechnung
        :return: 
        """
        print(self.mean)
    
    def __str__(self) -> str:
        return f"{self.sensorname}sensor, ID: {self.id} und Messwerten: {self.measurement}"
```

```python
# Anlegen der neuen Klassen
class ServoMotor():
    def __init__(self, name, soll_winkel, ist_winkel, leistungs_aufnahme=0) -> None:
        """
        Initialisieren

            name (_type_): _description_
            soll_winkel (_type_): _description_
            ist_winkel (bool): _description_
            leistungs_aufnahme (int, optional): _description_. Defaults to 0.
        """
        self.name = name
        self.soll_winkel = soll_winkel
        self.ist_winkel = ist_winkel
        self.leistungs_aufnahme = leistungs_aufnahme
    
    def zeige_current_angle(self):
        """
        Anzeige Ist-Winkel Motor
        """
        print(F"{self.name} bei {self.ist_winkel}°")
    
    def zeige_target_angle(self):
        """
        Anzeige Soll-Winkel Motor
        """
        print(F"{self.name} soll auf {self.soll_winkel}° stehen")
    
    def setze_target_angle(self, angle):
        """
        Setzt Soll-Winkel

            angle (_type_): _description_
        """
        self.soll_winkel = angle
        self.zeige_target_angle()
        
    def __str__(self) -> str:
        return f"Motor {self.name} steht auf {self.ist_winkel}°"

    
class RobotController():
    def __init__(self, name, motors, sensors) -> None:
        """Initialisieren RoboterController Klasse

            name (_type_): _description_
            motors (_type_): _description_
            sensors (_type_): _description_
        """
        self.name = name
        self.motors: list = motors
        self.sensors: list = sensors
    
    def add_sensor(self, sensor):
        """
        Erweiterung Sensorliste

            sensor (_type_): _description_
        """
        self.sensors.append(sensor)
    
    def add_motor(self, motor):
        """
        Erweiterung Motorliste

            motor (_type_): _description_
        """
        self.motors.append(motor)
            
    def resting_position(self):
        """
        Setzt Soll-Position 0
        """
        for motor in self.motors:
            motor.setze_target_angle(0)
    
    def __str__(self) -> str:
        motor_info = "\n".join([f"{motor.name}: Soll-Winkel = {motor.soll_winkel}°, Ist-Winkel = {motor.ist_winkel}°" for motor in self.motors])
        sensor_info = "\n".join([str(sensor) for sensor in self.sensors]) 
        return f"\n{self.name}-Motoren:\n{motor_info}\n\n{self.name}-Sensoren:\n{sensor_info}\n"
```

```python
# Beispiel, wie Sie diese Klassen verwenden können:
Temperatur = []
Feuchtigkeit = []
S1 = Sensor(1,"Temperatur", Temperatur)
S2 = Sensor(2,"Feuchtigkeit", Feuchtigkeit)

for i in range(15):
    S1.set_last_measurement(i)
    S2.set_last_measurement(2*i)

Beingelenk = ServoMotor("Beingelenk-Motor",20,0)
Armgelenk = ServoMotor("Armgelenk-Motor",50,0)

Beingelenk.zeige_current_angle()
Armgelenk.zeige_target_angle()
Armgelenk.setze_target_angle(45)

list_of_motors = []
list_of_sensors = []

Controller = RobotController("Roboter", list_of_motors, list_of_sensors)

Controller.add_sensor(S1)
Controller.add_sensor(S2)

Controller.add_motor(Beingelenk)
Controller.add_motor(Armgelenk)

print(Controller)

Controller.resting_position()

print(Controller)
```
